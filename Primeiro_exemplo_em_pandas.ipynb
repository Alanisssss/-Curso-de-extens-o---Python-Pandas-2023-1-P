```python
# Importando biblioteca pandas no python
import pandas as pd
combustiveis_df = pd.read_excel("ca-2021-02.xlsx")
```

```python
# Organizar melhor a tabela
display(combustiveis_df)
```

```python
# Comandos dataframe.shape e dataframe.describe()
print(combustiveis_df.shape)
# Mostra o número de linhas e colunas (linhas, colunas)

#para ter só as linhas
print(combustiveis_df.shape[0])
```

```python
#Quais são as colunas e que tipo de dados cada uma tem usa o dataframe.info()
display(combustiveis_df.info())
```

```python
#describe() mostra as estatisticas "mais basicas" que a gente precisa
print(combustiveis_df.describe())

#std é o desvio padrão
```

```python
#filtrar apenas por uma coluna
display(combustiveis_df['Revenda'])
```

```python
#Aqui criamos um novo dataframe apenas com as colunas que eu quero
ca_df = combustiveis_df[['Revenda','Produto','Municipio','Valor de Venda']]
display(ca_df)
```

```python
# exibe a 4a. linhas com .loc[]
display(ca_df.loc[3])
```

```python
#exibir da 10a. linhas ate a 20a.
display(ca_df.loc[9:19])
```

```python
#Criar um dataframe gas_df contendo 
#apenas as 4 colunas (Revenda, Municipio, Produto, Valor de Venda)
#somente com combustivel sendo GASOLINA e exibir na tela
#trazer valor da gasolina mais cara do brasil
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)
display(gas_df['Valor de Venda'].max())
```

```python
#Exibir qual é o posto e municipio da gasolina mais cara do país?
display(gas_df[['Revenda','Municipio','Valor de Venda']].max())
```
